```python
def finding_unique_words(file_path):
  text_file = open(file_path, 'r')
  text = text_file.read()

  #cleaning 
  text = text.lower() # lowering the words 
  words = text.split() #split the text into words separated by white space characters like single space, new line, tab, etc.
  words = [word.strip('.,!;()[]``*') for word in words] #method to strip the punctuation marks from the edges of words.
  words = [word.replace("s", '') for word in words]#replace 's with nothing, at the end of words.
  unique = []
  for word in words:
     if word not in unique:
         unique.append(word)

  #sorting the unique words
    #unique.sort()

  return len(unique)
  #print('Nombers of unique words are: ', len(unique))
```

```python
from aitextgen import aitextgen
ai = aitextgen(model = 'gpt2')
```

```python
from aitextgen.colab import mount_gdrive, copy_file_to_gdrive, copy_file_from_gdrive
mount_gdrive()
```

```python
from aitextgen.utils import GPT2ConfigCPU
config=GPT2ConfigCPU()
ai.train('Train',
         line_by_line=False,
         from_cache=False,
         num_steps=5000,
         generate_every=300,
         save_every=300,
         save_gdrive=True,
         learning_rate=1e-4,
         fp16=False,
         batch_size=3,
         config=config
         )
```

```python
ai.save()
```

```python
def input_text():
  text_input = list()
  num_of_sentences = input('Enter the number of sentences you want to put: ')
  for i in range(int(num_of_sentences)):
    text = input('Enter the text ' + str(i) + ':  ' )
    text_input.append(text)

  return  ' '.join([ i for i in text_input])
```

```python
#Calling the input text function for taking desired number of sentences.
text = input_text()
length=int(input('Input the size of your story:'))
```

```python
from_folder = None

for file in ["pytorch_model.bin", "config.json"]:
  if from_folder:
    copy_file_from_gdrive(file, from_folder)
  else:
    copy_file_from_gdrive(file)
```

```python
ai = aitextgen(model_folder=".")
```

```python
ai.generate(n = 3,
            prompt = text,
            max_length = length,
            temperature=.7,
            do_sample = True,
            return_as_list = False,
            seed = 42,
            top_p=0.9
            
 )
```
